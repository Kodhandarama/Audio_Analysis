```python
%matplotlib inline
from pathlib import Path
import librosa, librosa.display
import seaborn
import numpy as np
import numpy, scipy, matplotlib.pyplot as plt, librosa, IPython.display as ipd
```

```python
x, sr = librosa.load('/03_Balasarasamurali_Keeravani_Adi_OothukaduVenkatakavi.wav')
```

```python
print (x.shape)
print (sr)
```

```python
ipd.Audio(x, rate=sr)
```

```python
X = scipy.fft(x)
X_mag = numpy.absolute(X)
f = numpy.linspace(0, sr, len(X_mag)) # frequency variable
```

```python
plt.figure(figsize=(13, 5))
plt.plot(f, X_mag) # magnitude spectrum
plt.xlabel('Frequency (Hz)')
```

```python
plt.figure(figsize=(13, 5))
plt.plot(f[:5000], X_mag[:5000])
plt.xlabel('Frequency (Hz)')
```

```python
S = np.abs(librosa.stft(x, n_fft=4096))**2
chroma = librosa.feature.chroma_stft(S=S, sr=sr)
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chromagram')
plt.tight_layout()
plt.show()
```
